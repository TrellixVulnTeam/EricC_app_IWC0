```python
import findspark
findspark.init()

import pyspark
from pyspark.sql import SparkSession

spark = SparkSession.builder.getOrCreate()
df = spark.read.csv("nuclear_plants_small_dataset.csv", inferSchema=True, header=True)

def calcMin(data):
    print(data)
    minValue = None
    
    data.foreach(lambda x:
        print(x)
    )
    
    return minValue

normal_data = df.where(df.Status == "Normal").select(df.Power_range_sensor_1)

power_range_sensor_1_min_value = calcMin(normal_data)
print(power_range_sensor_1_min_value)
```
